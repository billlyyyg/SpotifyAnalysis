```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
#Create path way to data folder
datafolder = Path("data")
#Create path way to tracks data set
tracks_data = pd.read_csv(datafolder / "tracks.csv")
tracks_data.head()
```

```python
#check for null values

pd.isnull(tracks_data).sum()
```

```python
tracks_data.info()
```

```python
#find most popular songs
pop_sorted_tracks = tracks_data.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)

pop_sorted_tracks[:10]
```

```python
#display description of tracks
tracks_data.describe().transpose()
```

```python
#set date column to index
tracks_data.set_index("release_date", inplace = True)
#convert dates to datetime datatype
tracks_data.index = pd.to_datetime(tracks_data.index, format="mixed")
tracks_data.head()
```
